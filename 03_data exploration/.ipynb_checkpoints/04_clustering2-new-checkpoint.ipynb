```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
%matplotlib inline
```

```python
cltNum = {'c_':2, 'j_':3, 'ht_':14, 's_':4, 'cs_':15, 'e_':4}
flt = ['c_', 'j_', 'ht_', 's_', 'cs_', 'e_']
result2  = pd.read_csv('train'+str(105)+'.csv').drop(columns=['index', 'Year', 'County Name', 'Township Name', 'Village Name'])
for j in flt:
    for i in range(105,110):
        df = pd.read_csv('train'+str(i)+'.csv').drop(columns=['index', 'Year', 'County Name', 'Township Name', 'Village Name'])
        df1 = df.filter(regex='^'+j)
        X = df1
        
        if i == 105:
            final_model = KMeans(n_clusters=cltNum[j], random_state=42)
            final_model.fit(X)
                
        # Fit model to data
        
        labels = final_model.labels_
        sil = silhouette_score(X, labels)
        db = davies_bouldin_score(X, labels)
        print(str(i)+"|"+j+" |db index:", db)
        print(str(i)+"|"+j+" |silhouette_scores:", sil)
        result2[str(i)+"|"+j] = list(final_model.predict(X))
        
        
result2 = result2.iloc[:,65:]
result2
```

```python
result = result2.copy()
for i in range(105,110):
    df = pd.read_csv('train'+str(i)+'.csv').drop(columns=['index', 'Year', 'County Name', 'Township Name', 'Village Name'])
    X = df.iloc[:, :-1]
    
    print("This is year:", i)
    if i ==105:
        final_model3 = KMeans(n_clusters=3, random_state=42)
        final_model3.fit(X)
        final_model7 = KMeans(n_clusters=7, random_state=42)
        final_model7.fit(X)
        final_model13 = KMeans(n_clusters=13, random_state=42)
        final_model13.fit(X)
        
    # all features

    # Fit model to data
    
    labels = final_model3.labels_
    sil = silhouette_score(X, labels)
    db = davies_bouldin_score(X, labels)
    print(str(i)+"|all features 3|db index:", db)
    print(str(i)+"|all features 3|silhouette_scores:", sil)
    result[str(i)+"|"+'kmeans_3'] = list(final_model3.predict(X))
    

    # Fit model to data
    
    labels = final_model7.labels_
    sil = silhouette_score(X, labels)
    db = davies_bouldin_score(X, labels)
    print(str(i)+"|all features 7|db index:", db)
    print(str(i)+"|all features 7|silhouette_scores:", sil)
    result[str(i)+"|"+'kmeans_7'] = list(final_model7.predict(X))
    
    
    # Fit model to data
    
    labels = final_model13.labels_
    sil = silhouette_score(X, labels)
    db = davies_bouldin_score(X, labels)
    print(str(i)+"|all features 13|db index:", db)
    print(str(i)+"|all features 13|silhouette_scores:", sil)
    result[str(i)+"|"+'kmeans_13'] = list(final_model13.predict(X))
    
    # 
```

```python
for i in range(105,110):
    #save file
    result.filter(regex=str(i)+'|V_ID').to_csv('0806_'+str(i)+'.csv')
result.to_csv('combine_clustering_0806.csv')
```
